```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as ipw
from ipywidgets import* 
from IPython.display import display 
```

### interact decorator 

```python
#delete data below and enter your data in x_exp and y_exp,keep the brackets
x_exp = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y_exp = [100,90,80,70,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
```

```python
df=pd.DataFrame({'time':x_exp,'speed':y_exp})
df.head()
```

## choose poly degree

```python
@interact()
def poly_degree(poly_degree_n=list(range(1,10))):
    plt.figure(figsize=(12,6))
    plt.style.use('fivethirtyeight')
    mymodel = np.poly1d(np.polyfit(df.time, df.speed, poly_degree_n))
    print('fitting_model:')
    print(mymodel)
    print(65*'*')
    plt.scatter(df.time,df.speed,marker='d',c='navy',label='measured_data',s=150)
    plt.scatter(np.arange(1,np.max(x_exp),0.5),mymodel(np.arange(1,np.max(x_exp),0.5)),
                s=100,marker='D',c='lime',label='fitted_data')
                
    plt.legend(loc='upper right')
    plt.grid(ls='--',color='grey')
    plt.xlabel(df.columns[0],c='blue')
    plt.ylabel(df.columns[1],c='blue')
    plt.xlim(0,df.time.max()+1.)
    plt.xticks(np.arange(0,df.time.max()+1.1,2))
    print()
    print('values of fit with more precicion')
    display(mymodel)
print('program run successfully')
```
